```python
# 可视化tsne
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

# 假设 loaded_data 是你刚刚 torch.load 出来的列表
# features, labels = loaded_data 

def visualize_features(features, labels, samples_per_class=0):
    """
    :param features: Tensor, 形状 [N, D]
    :param labels: Tensor, 形状 [N]
    :param samples_per_class: 每个类别采样的数量
    """
    
    # 1. 预处理：将 Tensor 转换为 Numpy 数组
    # .detach() 去掉梯度, .cpu() 移到 CPU
    if isinstance(features, torch.Tensor):
        features = features.detach().cpu().numpy()
    if isinstance(labels, torch.Tensor):
        labels = labels.detach().cpu().numpy()

    print(f"原始数据形状: Features={features.shape}, Labels={labels.shape}")

    # 2. 采样：每个类别挑选 30 个样本
    unique_classes = np.unique(labels)
    selected_indices = []

    for cls in unique_classes:
        # 找到属于该类别的所有索引
        cls_indices = np.where(labels == cls)[0]
        
        # 如果该类样本数大于 30，随机选 30 个；否则全选
        if len(cls_indices) > samples_per_class:
            # replace=False 表示不重复抽取
            chosen = np.random.choice(cls_indices, samples_per_class, replace=False)
        else:
            chosen = cls_indices
            
        selected_indices.extend(chosen)

    # 提取筛选后的数据
    X_sample = features[selected_indices]
    y_sample = labels[selected_indices]

    print(f"采样后数据形状: Features={X_sample.shape}, Labels={y_sample.shape}")

    # 3. 降维：使用 t-SNE (适合观察聚类效果)
    print("正在进行 t-SNE 降维 (可能需要几秒钟)...")
    tsne = TSNE(n_components=2, init='pca', learning_rate='auto', random_state=42)
    X_embedded = tsne.fit_transform(X_sample)

    # 4. 绘图
    plt.figure(figsize=(10, 8))
    
    # 使用 scatter 绘制散点图
    # c=y_sample 根据标签上色, cmap='jet' 使用高对比度色谱
    scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], 
                          c=y_sample, cmap='jet', alpha=0.7, s=30)
    
    # 添加颜色条 (Legend)
    cbar = plt.colorbar(scatter)
    cbar.set_label('Classes')
    
    plt.title(f't-SNE Visualization ({samples_per_class} samples/class)')
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.grid(True, alpha=0.3)
    plt.show()

# --- 调用函数 ---
# 确保 features 和 labels 已经被赋值
# features, labels = torch.load("你的路径.pth", map_location='cpu') 
visualize_features(features, labels, samples_per_class=60)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def visualize_comparison(features, labels, fake_features, 
                           real_samples_per_class=60, 
                           fake_samples_count=300, 
                           fake_label_id=6):
    """
    绘制两幅对比图：
    1. 仅展示真实样本分布
    2. 展示真实样本 + 伪造样本分布
    (基于同一坐标系，方便对比)
    """
    
    # --- 1. 数据预处理 ---
    if isinstance(features, torch.Tensor):
        features = features.detach().cpu().numpy()
    if isinstance(labels, torch.Tensor):
        labels = labels.detach().cpu().numpy()
    if isinstance(fake_features, torch.Tensor):
        fake_features = fake_features.detach().cpu().numpy()

    # --- 2. 真实数据采样 ---
    unique_classes = np.unique(labels)
    real_indices = []
    for cls in unique_classes:
        cls_indices = np.where(labels == cls)[0]
        if len(cls_indices) > real_samples_per_class:
            chosen = np.random.choice(cls_indices, real_samples_per_class, replace=False)
        else:
            chosen = cls_indices
        real_indices.extend(chosen)

    X_real_sample = features[real_indices]
    y_real_sample = labels[real_indices]
    num_real = len(X_real_sample) # 记录真实样本数量

    # --- 3. 伪造数据采样 ---
    total_fake = len(fake_features)
    if total_fake > fake_samples_count:
        fake_idx = np.random.choice(total_fake, fake_samples_count, replace=False)
        X_fake_sample = fake_features[fake_idx]
    else:
        X_fake_sample = fake_features
    
    y_fake_sample = np.full(len(X_fake_sample), fake_label_id)

    print(f"采样统计 -> 真实样本: {num_real}, 伪造样本: {len(X_fake_sample)}")

    # --- 4. 拼接并统一进行 t-SNE ---
    # 注意：我们把它们拼在一起做 t-SNE，这样坐标系是固定的，方便肉眼对比
    X_combined = np.concatenate((X_real_sample, X_fake_sample), axis=0)
    y_combined = np.concatenate((y_real_sample, y_fake_sample), axis=0)

    print("正在进行联合 t-SNE 降维（请稍候）...")
    tsne = TSNE(n_components=2, init='pca', learning_rate='auto', random_state=42)
    X_embedded = tsne.fit_transform(X_combined)

    # 拆分回真实部分和伪造部分
    X_embedded_real = X_embedded[:num_real]
    y_embedded_real = y_combined[:num_real]
    
    # --- 5. 绘图 (双子图) ---
    fig, axes = plt.subplots(1, 2, figsize=(20, 9))
    
    # 为了保持颜色一致，我们需要固定颜色范围
    vmin = min(y_combined)
    vmax = max(y_combined)

    # === 左图：只看真实样本 ===
    scatter1 = axes[0].scatter(X_embedded_real[:, 0], X_embedded_real[:, 1], 
                               c=y_embedded_real, cmap='jet', 
                               vmin=vmin, vmax=vmax, alpha=0.7, s=40)
    axes[0].set_title("Original Feature Space (Real Only)")
    axes[0].set_xlabel("t-SNE Dim 1")
    axes[0].set_ylabel("t-SNE Dim 2")
    axes[0].grid(True, alpha=0.3)
    
    # === 右图：混合样本 ===
    # 先画真实样本（做底）
    axes[1].scatter(X_embedded_real[:, 0], X_embedded_real[:, 1], 
                    c=y_embedded_real, cmap='jet', 
                    vmin=vmin, vmax=vmax, alpha=0.4, s=40) # alpha调低一点，让背景淡一点
    
    # 再画伪造样本（叠加在上面，或者混在一起）
    # 提取伪造部分的坐标
    X_embedded_fake = X_embedded[num_real:]
    y_embedded_fake = y_combined[num_real:]
    
    scatter2 = axes[1].scatter(X_embedded_fake[:, 0], X_embedded_fake[:, 1], 
                               c=y_embedded_fake, cmap='jet', 
                               vmin=vmin, vmax=vmax, alpha=0.9, s=50, marker='*') # 用星号标记伪造样本
    
    axes[1].set_title(f"Mixed Feature Space (with Generated Class {fake_label_id})")
    axes[1].set_xlabel("t-SNE Dim 1")
    axes[1].grid(True, alpha=0.3)

    # 添加颜色条 (共用)
    cbar = fig.colorbar(scatter2, ax=axes, orientation='horizontal', fraction=0.05, pad=0.1)
    cbar.set_label(f'Classes (0-{fake_label_id-1}: Real, {fake_label_id}: Generated)')

    plt.show()

# --- 调用示例 ---
visualize_comparison(features, labels, fake_features, real_samples_per_class=100, fake_samples_count=500)
```
